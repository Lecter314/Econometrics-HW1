```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("TV_dataset_preprocessing1.1.csv")
TV_dataset_preprocessing = df.copy()
TV_dataset_preprocessing.head()
```

```python
fig, ax =plt.subplots(1, 3, constrained_layout=True, figsize=(16, 8))

axsub = TV_dataset_preprocessing['Brand'].value_counts().plot(kind='bar', ax=ax[0])
axsub.set_ylabel('frequency')
axsub = TV_dataset_preprocessing['Resolution'].value_counts().plot(kind='bar', ax=ax[1])
axsub = TV_dataset_preprocessing['Operating System'].value_counts().plot(kind='bar', ax=ax[2])

fig.suptitle('bar charts for categorical variables')
plt.show()
```

```python
# lots of observations are from Others
fig, ax =plt.subplots(1, 3, constrained_layout=True, figsize=(16, 8))

axsub = TV_dataset_preprocessing['Brand'].value_counts().plot(kind='box', ax=ax[0])
axsub.set_ylabel('frequency')
axsub = TV_dataset_preprocessing['Resolution'].value_counts().plot(kind='box', ax=ax[1])
axsub = TV_dataset_preprocessing['Operating System'].value_counts().plot(kind='box', ax=ax[2])

fig.suptitle('Box plot for categorical variables')
plt.show()
```

```python
sns.jointplot(data=TV_dataset_preprocessing, x='Size ', y='Rating', height=12, ratio=2, marginal_ticks=True)
plt.title('Relations between Size and Rating', fontsize=18)
TV_dataset_preprocessing[['Size ', 'Rating']].corr().apply(lambda x: x.round(4))
```

```python
sns.jointplot(data=TV_dataset_preprocessing, x='Selling Price', y='Rating', height=12, ratio=2, marginal_ticks=True)
plt.title('Relations between Selling Price and Rating', fontsize=18)
TV_dataset_preprocessing[['Selling Price', 'Rating']].corr().apply(lambda x: x.round(4))
```

```python
sns.jointplot(data=TV_dataset_preprocessing, x='Original Price', y='Rating', height=12, ratio=2, marginal_ticks=True)
plt.title('Relations between Original Price and Rating', fontsize=18)
TV_dataset_preprocessing[['Original Price', 'Rating']].corr().apply(lambda x: x.round(4))
```

```python
# see? multicollinearity
sns.jointplot(data=TV_dataset_preprocessing, x='Original Price', y='Selling Price', height=12, ratio=2, marginal_ticks=True)
plt.title('Relations between Original Price and Selling Price', fontsize=18)
TV_dataset_preprocessing[['Original Price', 'Selling Price']].corr().apply(lambda x: x.round(4))
```

```python
# see? multicollinearity
sns.jointplot(data=TV_dataset_preprocessing, x='Size ', y='Selling Price', height=12, ratio=2, marginal_ticks=True)
plt.title('Relations between Size and Selling Price', fontsize=18)
TV_dataset_preprocessing[['Size ', 'Selling Price']].corr().apply(lambda x: x.round(4))
```

#### prepare the initial dataset for R

```python
# actually, selecting the needed conlums
TV_dataset_output = TV_dataset_preprocessing[["Brand", "Resolution", "Size ", "Selling Price", "Original Price", "Operating System", "Rating"]]
TV_dataset_output = TV_dataset_output.rename(columns={"Size ":"Size"})
```

```python
brand_dict = {"Others": 0, "SAMSUNG": 2, "LG": 3, "SONY": 4, "TCL": 5}
resolution_dict = {"Ultra HD LED": 0, "Full HD LED": 1, "HD LED": 2}
os_dict = {"not Android": 0, "Android": 1}
TV_dataset_output = TV_dataset_output.replace({"Brand": brand_dict})
TV_dataset_output = TV_dataset_output.replace({"Resolution": resolution_dict})
TV_dataset_output = TV_dataset_output.replace({"Operating System": os_dict})
TV_dataset_output.head()
```

```python
# rename columns
TV_dataset_output.columns = ["brand", "reso", "size", "sellp", "origp", "os", "rate"]
```
